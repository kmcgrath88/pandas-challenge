```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools.csv"
student_data_to_load = "Resources/students.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.to_csv("Resources/MergedData.csv")
```

```python
#Find column names
#print(school_data_complete.columns)

#Calculate the total number of schools 
TotalSchools = school_data['school_name'].tolist()
print(len(TotalSchools))

#Calculate the total number of students
TotalStudents = school_data['size'].tolist()
print(TotalStudents)
SumTotalStudents = sum(TotalStudents)
print(f'{SumTotalStudents:,}')


#SchoolSummary = {'Total Schools': [TotalSchools[0:]], 'Total Students': [TotalStudents[0:]]}
#SchoolSummaryDF = pd.DataFrame(SchoolSummary).transpose()
#SchoolSummaryDF.columns=['Total Schools', 'Total Students']
#SchoolSummaryDF
```

```python
#Calculate the total budget
Budget = school_data["budget"].tolist()
TotalBudget = sum(Budget)
print(f'${TotalBudget:,.2f}')
```

```python
#Calculate the average math score
AvgMathScore = school_data_complete["math_score"].mean()
print(AvgMathScore)

#Calculate the average reading score
AvgReadScore = school_data_complete["reading_score"].mean()
print(AvgReadScore)
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
MathOver70 = len(school_data_complete.loc[school_data_complete["math_score"]>=70, "math_score"])
MathPercentPass = MathOver70/SumTotalStudents *100
print(MathPercentPass)

#Calculate the percentage of students with a passing reading score (70 or greater)
ReadOver70 = len(school_data_complete.loc[school_data_complete["reading_score"]>=70, "reading_score"])
ReadPercentPass = ReadOver70/SumTotalStudents *100
print(ReadPercentPass)

#Calculate the percentage of students who passed math and reading (% Overall Passing)
TotalPercentPassed = school_data_complete.loc[(school_data_complete["reading_score"]>=70) & 
                        (school_data_complete["math_score"]>=70), :]
PassedBoth = len(TotalPercentPassed)/SumTotalStudents * 100
print(PassedBoth)
```

```python
#Create a dataframe to hold the above results
DistSummary = {"Total Schools" : [len(TotalSchools)], "Total Students" : [f'{SumTotalStudents:,}'],
                     "Total Budget" : [f'${TotalBudget:,.2f}'], "Average Math Score" : [AvgMathScore],
                     "Average Reading Score" : [AvgReadScore], "% Passing Math" : [MathPercentPass],
                     "% Passing Reading" : [ReadPercentPass], "% Overall Passing" : [PassedBoth]}
DistrictSummaryDF = pd.DataFrame(DistSummary)

DistrictSummaryDF

#DS = districtsummaryDF.to_csv("District Summary.csv")

#DS1 = "District Summary.csv"
#DS2 = pd.read_csv(DS1)
#DS2

#Optional: give the displayed data cleaner formatting ---- come back to
```

```python
#GroupbySchool
SchoolGroup = school_data_complete.groupby(['school_name'])

#Total Students
StudentTotal = SchoolGroup['student_name'].count()

#Type
Type = SchoolGroup['type'].unique() #------????

#Budget
NewBudget = SchoolGroup['budget'].max()

#PerStudentBudget
PerStudentBudget = NewBudget/StudentTotal

#Average Math Score
MathScore = SchoolGroup['math_score'].mean()

#Average Reading Score
EngScore = SchoolGroup['reading_score'].mean()
```

```python
#% Passing Math
#Created new DF for math scores over 70
#Regrouped data by school name using new DF, found the total number, calculated percent
MathPass = school_data_complete[(school_data_complete['math_score']>=70)]
SchoolMathGroup = MathPass.groupby(['school_name'])
MathPassSum = SchoolMathGroup['math_score'].count()
MathPassPercent = (MathPassSum/StudentTotal) * 100

#% Passing Reading
EngPass = school_data_complete[(school_data_complete['reading_score']>=70)]
SchoolEngGroup = EngPass.groupby(['school_name'])
EngPassSum = SchoolEngGroup['reading_score'].count()
EngPassPercent = (EngPassSum/StudentTotal) * 100

#% Overall Passing (The percentage of students that passed math and reading.)
MathEngPass = school_data_complete[(school_data_complete['reading_score']>=70)&(school_data_complete['math_score']>=70)]
SchoolCombGroup = MathEngPass.groupby(['school_name'])
CombPassSum = SchoolCombGroup['math_score'].count()
CombPassPercent = (CombPassSum/StudentTotal) * 100

```

```python
#School Name, School Type, Total Students, Total School Budget
SchoolSummary = {'School Type': Type, 'Total Students': StudentTotal, 'Total School Budget': NewBudget, 
                 'Per Student Budget': PerStudentBudget, 'Average Math Score': MathScore,
                 'Average English Score': EngScore, '% Passing Math': MathPassPercent,
                 '% Passing Reading': EngPassPercent, '% Overall Passing': CombPassPercent}
SchoolSummaryDF = pd.DataFrame(SchoolSummary) 


#SchoolSummaryDF["School Type"] = SchoolSummaryDF["School Type"].str.strip('[]')
SchoolSummaryDF["Total Students"] = SchoolSummaryDF["Total Students"].map("{:,}".format)
SchoolSummaryDF["Total School Budget"] = SchoolSummaryDF["Total School Budget"].map("${:,.2f}".format)
SchoolSummaryDF["Per Student Budget"] = SchoolSummaryDF["Per Student Budget"].map("${:,.2f}".format)
SchoolSummaryDF["Average Math Score"] = SchoolSummaryDF["Average Math Score"].map("{:.2f}".format)
SchoolSummaryDF["Average English Score"] = SchoolSummaryDF["Average English Score"].map("{:.2f}".format)
#Create a dataframe to hold the above results
SchoolSummaryDF.head()
```

```python
Top5Overall_df = SchoolSummaryDF.sort_values("% Overall Passing", ascending = False)
Top5Overall_df.head()
```

```python
Worst5Overall_df = SchoolSummaryDF.sort_values("% Overall Passing", ascending = True)
Worst5Overall_df.head()
```

```python
#Create a table that lists the average Math Score for students of each grade level (9th, 10th, 11th, 12th) at each school.
#Create a pandas series for each grade.
Math9th_series = school_data_complete.loc[(school_data_complete['grade'] == '9th')]
Math9th_series = Math9th_series[['school_name','grade','math_score']]
Math9th_series = Math9th_series.rename(columns = {'school_name' : ' ' , 'math_score' : '9th'})
Math9th_series = Math9th_series.groupby([' ']).mean()

Math10th_series = school_data_complete.loc[(school_data_complete['grade'] == '10th')]
Math10th_series = Math10th_series[['school_name','grade','math_score']]
Math10th_series = Math10th_series.rename(columns = {'school_name' : ' ' , 'math_score' : '10th'})
Math10th_series = Math10th_series.groupby([' ']).mean()

Math11th_series = school_data_complete.loc[(school_data_complete['grade'] == '11th')]
Math11th_series = Math11th_series[['school_name','grade','math_score']]
Math11th_series = Math11th_series.rename(columns = {'school_name' : ' ' , 'math_score' : '11th'})
Math11th_series = Math11th_series.groupby([' ']).mean()

Math12th_series = school_data_complete.loc[(school_data_complete['grade'] == '12th')]
Math12th_series = Math12th_series[['school_name','grade','math_score']]
Math12th_series = Math12th_series.rename(columns = {'school_name' : ' ' , 'math_score' : '12th'})
Math12th_series = Math12th_series.groupby([' ']).mean()

#Combine the series into a dataframe
MergedMath_df = Math9th_series.merge(Math10th_series,on=' ').merge(Math11th_series,on=' ').merge(Math12th_series,on=' ')
MergedMath_df.head()
```

```python
#Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.
#Create a pandas series for each grade. Hint: use a conditional statement.
Read9th_series = school_data_complete.loc[(school_data_complete['grade'] == '9th')]
Read9th_series = Read9th_series[['school_name','grade','reading_score']]
Read9th_series = Read9th_series.rename(columns = {'school_name' : ' ' , 'reading_score' : '9th'})
Read9th_series = Read9th_series.groupby([' ']).mean()

Read10th_series = school_data_complete.loc[(school_data_complete['grade'] == '10th')]
Read10th_series = Read10th_series[['school_name','grade','reading_score']]
Read10th_series = Read10th_series.rename(columns = {'school_name' : ' ' , 'reading_score' : '10th'})
Read10th_series = Read10th_series.groupby([' ']).mean()

Read11th_series = school_data_complete.loc[(school_data_complete['grade'] == '11th')]
Read11th_series = Read11th_series[['school_name','grade','reading_score']]
Read11th_series = Read11th_series.rename(columns = {'school_name' : ' ' , 'reading_score' : '11th'})
Read11th_series = Read11th_series.groupby([' ']).mean()

Read12th_series = school_data_complete.loc[(school_data_complete['grade'] == '12th')]
Read12th_series = Read12th_series[['school_name','grade','reading_score']]
Read12th_series = Read12th_series.rename(columns = {'school_name' : ' ' , 'reading_score' : '12th'})
Read12th_series = Read12th_series.groupby([' ']).mean()

#Combine the series into a dataframe
MergedRead_df = Read9th_series.merge(Read10th_series,on=' ').merge(Read11th_series,on=' ').merge(Read12th_series,on=' ')
MergedRead_df.head()
```
